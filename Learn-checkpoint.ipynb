```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH_df = 'D:/Python/Datasets/WA_Fn-UseC_-Accounts-Receivable.csv'
df = pd.read_csv(PATH_df)
```

```python
df
print("{:,} order lines".format(len(df)))
df.head()
```

```python
# InvoiceAmout/customerID
df_par = pd.DataFrame(df.groupby(['customerID'])['InvoiceAmount'].sum())
df_par.columns = ['InvoiceAmount']

# Sort Values
df_par.sort_values(['InvoiceAmount'], ascending = False, inplace = True)
df_par.reset_index(inplace = True)

# Cumulative Sum 
df_par['CumSum'] = df_par['InvoiceAmount'].cumsum()

# % CumSum
df_par['%CumSum'] = (100 * df_par['CumSum']/df_par['InvoiceAmount'].sum())

# % Customer
df_par['%customerID'] = (100 * (df_par.index + 1).astype(float)/(df_par.index.max() + 1))


# 80% Volume
df_par80 = df_par[df_par['%CumSum'] > 80].copy()
perc_cust80 = df_par[df_par['%CumSum'] > 80]['%customerID'].min()
perc_sum80 = df_par[df_par['%CumSum'] > 80]['%CumSum'].min()

# 20% Customer
df_cust20 = df_par[df_par['%customerID'] > 20].copy()
perc_cust20 = df_cust20['%customerID'].min()
perc_sum20 = df_cust20['%CumSum'].min()


print("Pareto Analysis for {:,} unique customerID".format(len(df_par)))
df_par.head(5)
```

```python
ax = df_par.plot(x='%customerID', y='%CumSum', figsize = (15,7), color = 'tab:blue', legend = False, linewidth=3)
plt.xlabel('(%) customerID',fontsize=15)
plt.ylabel('(%) InvoiceAmount',fontsize=15)
plt.title('Pareto Analysis by Inoivces Amount', fontsize = 15)

# 80% Volume
ax.axhline(perc_sum80 , color="tab:green", linestyle="--", linewidth = 2.0)
ax.axvline(perc_cust80, color="tab:green", linestyle="--", linewidth = 2.0)
# 20% Customer
ax.axhline(perc_sum20 , color="tab:orange", linestyle="--", linewidth = 2.0)
ax.axvline(perc_cust20, color="tab:orange", linestyle="--", linewidth = 2.0)
plt.show()
```
